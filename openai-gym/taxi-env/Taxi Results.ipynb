```python
import json
import datetime
import matplotlib.pyplot as plt
```

```python
def loadJson(file):
    jsonFile = open(file)
    return json.load(jsonFile)
```

```python
def timestamp_to_time(timestamp):
    return datetime.datetime.fromtimestamp(
        int(timestamp)
    )
```

```python
def getTimestamps(jsonData): 
    start_timestamp = timestamp_to_time(jsonData['initial_reset_timestamp'])
    timestamps = [timestamp_to_time(x) - start_timestamp for x in jsonData['timestamps']]
    print(timestamp_to_time(jsonData['timestamps'][1]))
    return timestamps
```

```python
policyIterationData = loadJson("/home/joistick11/git/reinforcement-learning/openai-gym/taxi-env/res/policy_iteration/monitor/openaigym.episode_batch.0.10633.stats.json")
policy_episode_rewards = policyIterationData['episode_rewards']
policy_timestemps = getTimestamps(policyIterationData)
```

```python
valueIterationData = loadJson("/home/joistick11/git/reinforcement-learning/openai-gym/taxi-env/res/value_iteration/monitor/openaigym.episode_batch.0.10559.stats.json")
value_episode_rewards = valueIterationData['episode_rewards']
value_timestemps = getTimestamps(valueIterationData)
```

```python
plt.figure(figsize = (15, 8))
plt.plot([i for i in range(0, len(value_episode_rewards))], value_episode_rewards, linestyle='', marker='o', color="r", label="Value Iteration")
plt.plot([i for i in range(0, len(value_episode_rewards))], policy_episode_rewards, linestyle='', marker='o', color="b", label="Policy Iteration")
plt.xlim(0, 200)
plt.ylim(0, 20)
plt.legend()
plt.show()
```
